```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epsilon = 2e-16
x_range = np.arange(1e-5, 1e0+1e-5, 1e-5)
```

```python
fig, ax = plt.subplots()
ax.loglog(x_range, 60480*x_range**6, label = 'Truncation error')
ax.loglog(x_range, 5*epsilon - 7*epsilon/20.0*x_range**2, label = 'Rounding error')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('error')
```
