```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.exp(x/2 - 1) - x

def secant(f, x0, x1, t):
    
    x_km1 = x0
    x_k = x1
    
    # allocate space for estimates at each step
    x_i = np.ones((t, 1))
    
    for i in range(t):
        print('iteration number {}'.format(i))
        x_kp1 = x_k - f(x_k)*(x_k - x_km1)/(f(x_k) - f(x_km1))
        x_i[i] = x_kp1
        # substitute
        x_km1 = x_k
        x_k = x_kp1
    
    return(x_kp1, x_i)
```

```python
t = 17
x0 = 20
x1 = 15
xt, x_i = secant(f = f, x0 = x0, x1 = x1, t = t)
```

```python
err = abs(x_i - xt)
step_err = abs(err[1::] - err[0:-1])

fig, axes = plt.subplots(1,2)
axes[0].semilogy(list(range(t)), err)
axes[1].plot(list(range(t)[1:]), step_err)
axes[0].set_ylabel(r'error $|x_t - x_T|$')
axes[1].set_ylabel(r'error $|x_t - x_{t-1}|$')
axes[0].set_xlabel('iteration')
axes[1].set_xlabel('iteration')
fig.tight_layout()
```

```python
# a trickier function
def g(x):
    return x**4 - 4*x**2 + 4

t = 25
x0 = 1.413
x1 = 1.412
xt, x_i = secant(f = g, x0 = x0, x1 = x1, t = t)

# error and plot
err = abs(x_i - xt)
step_err = abs(err[1::] - err[0:-1])

fig, axes = plt.subplots(1,2)
axes[0].semilogy(list(range(t)), err)
axes[1].plot(list(range(t)[1:]), step_err)
axes[0].set_ylabel(r'error $|x_t - x_T|$')
axes[1].set_ylabel(r'error $|x_t - x_{t-1}|$')
axes[0].set_xlabel('iteration')
axes[1].set_xlabel('iteration')
fig.tight_layout()
```
