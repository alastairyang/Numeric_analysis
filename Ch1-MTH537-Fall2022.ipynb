```python
#Exercise 1.4

def f(h):
    return 1 - h**2/2

fig,ax = plt.subplots(1,3,figsize=(12,3))

h_s = linspace(0,1,1000)
ax[0].plot(h_s,f(h_s),'--',label="$1 - h^2/2$")
ax[0].plot(h_s,cos(h_s),'k',label="$cos(h)$")
ax[0].set_xlabel('h')
ax[0].legend();
ax[1].plot(h_s,cos(h_s)-f(h_s),'--',label="$cos(h_s)-[1 - h^2/2]$")
ax[1].legend();
ax[1].set_xlabel('h')
ax[2].plot(log(h_s),log(cos(h_s)-f(h_s)),'--',label="$log(cos(h_s)-[1 - h^2/2])$")
ax[2].legend();
ax[2].set_xlabel('log(h)')
plt.tight_layout()
```

```python
# The slope in the above plot is approximately 4, since the error is O(h^4)
```

```python
#Exercise 1.5

def f(x_n):
    x_n1 = (x_n**2 + 4)/(2*x_n)
    return x_n1

N=5
x = zeros(N)
x[0] = 3
for n in range(N-1):
    x[n+1] = f(x[n])
```

```python
x
```

```python
fig,ax = plt.subplots(1,3,figsize=(12,3))

ax[0].plot(arange(N),x,'--',label="$x_n$")
ax[0].legend();
ax[0].set_xlabel('n')
ax[1].plot(arange(N),x-2,'--',label="$x_n-2$")
ax[1].legend();
ax[1].set_xlabel('n')
ax[2].plot(arange(N),log(x-2),'--',label="$log(x_n-2)$")
ax[2].legend();
ax[2].set_xlabel('n')
plt.tight_layout()
```

```python
# Check convergence rate
```

```python
(x[1:]-2) / (x[:-1]-2) # no, not linear
```

```python
(x[1:]-2) / (x[:-1]-2)**2 # yes, quadratic with constant 1/4
```

```python
#Exercise 1.6

def h(k):
    x_k = 1/ ( 2**(k**2) )
    return x_k

K=10
x = zeros(K)
for k in range(K):
    x[k] = h(k)
```

```python
x
```

```python
fig,ax = plt.subplots(1,3,figsize=(12,3))

ax[0].plot(arange(K),x,'--',label="$x_k$")
ax[0].legend();
ax[0].set_xlabel('k')
ax[1].plot(arange(K),log(x),'--',label="$log(x_k)$")
ax[1].legend();
ax[1].set_xlabel('k')
ax[2].plot(log(arange(K)),log(x),'--',label="$log(x_k)$")
ax[2].legend();
ax[2].set_xlabel('log(k)')
plt.tight_layout()
```
