## Ch. 2, Sec. 2: Bisection method

```python
#Exercise 2.1

def f(x):
    return exp(x) + x

x_s = linspace(-1,0,100)
plot(x_s,f(x_s),label="$e^x + x$")
plot(x_s,[0]*len(x_s),'k:',label="0")
legend();
```

```python
def bisection(f,a,b,eps):
    if f(a)*f(b)>=0:
        print('agh!')
        return
    
    xs = []
    while (b-a)/2 > eps:
        x = (a+b)/2
        xs.append(x)
        print(x)
        if f(x)*f(a)>0:
            a = x
            b = b
        else:
            b = x
            a = a
            
    return x,xs
```

```python
a=-1
b=0
eps = 10**-14
root,xs = bisection(f,a,b,eps)
```

```python
x_s = linspace(-1,0,100)
plot(x_s,f(x_s),label="$e^x + x$")
plot(x_s,[0]*len(x_s),'k:',label="0")
plt.scatter(root,0)
legend()
plt.xlabel('$x$')
plt.ylabel('$e^x + x$')
```

```python
error = abs(array(xs[:-1]) - xs[-1])
plot(log(error))
```

## Ch. 2, Sec. 3: Fixed Point Method

```python
#Exercise 2.3
def g(x): return (x+1)/2
x=0 # initial condition must be close to actual root
for i in range (15):
    print(i,x)
    x = g(x)
```

```python
#Exercise 2.4
def g(x): return  - x**3/6 + x**5/120

def dg(x): return  - 3*x**2 /6 + 5*x**4/120

print('The roots (i.e., fixed points) are:')
print((sqrt(0)),(sqrt(720)),(-sqrt(720)) )

print('\nbut which are attracting fixed points?')
print(dg(sqrt(0)),dg(sqrt(720)),dg(-sqrt(720)) )

xs = linspace(-5,5,100)
plot(xs,g(xs),'k')
plot(xs,dg(xs),':')
legend()
```
