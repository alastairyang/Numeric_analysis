```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epsilon = 2e-16
x_range = np.arange(1e-5, 1e1+1e-5, 1e-5)
```

```python
fig, ax = plt.subplots()
ax.loglog(x_range, x_range**6/60480, label = 'Truncation error')
ax.loglog(x_range, 6*(1-epsilon)/x_range**2, label = 'Rounding error')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('error')
ax.set_title('HW1: Problem 3, Donglai Yang')
```
